Matrix symbols
SymPy can also operate on matrices of symbolic dimension ( n×m ). MatrixSymbol("M", n, m) creates a matrix  M  of shape  n×m .

```python
n, m = symbols('n m', integer=True)
M = MatrixSymbol("M", n, m)
b = MatrixSymbol("b", m, 1)

M*b, (M*b).shape, Transpose(M*b), Transpose(M*b).doit()
```

Indexed symbols can be created with IndexedBase and Idx. IndexedBase("A") represents an array A and Idx('i') represents an index i.

You can also use symbols('i') instead of Idx('i').

```python
A = IndexedBase("A")
i = Idx('i')

A[i], solve(Eq(x**2, 4), x), solve(x**2 + 3*x - 3, x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Maak een figuur en een 3D-as
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Definieer de parameters van de bol
radius = (sqrt(17))  # straal van de bol
u = np.linspace(0, 2 * np.pi, 100)  # hoek in de xy-vlak (theta)
v = np.linspace(0, np.pi, 100)  # hoek van z-as (phi)

# Sferische coördinaten omzetten naar cartesische coördinaten
x = radius * np.outer(np.cos(u), np.sin(v))
y = radius * np.outer(np.sin(u), np.sin(v))
z = radius * np.outer(np.ones(np.size(u)), np.cos(v))

# Plot de bol
ax.plot_surface(x, y, z, color='b', rstride=10, cstride=10, alpha=0.6)

# Stel titels en labels in
ax.set_title("3D Bol")
ax.set_xlabel("X-as")
ax.set_ylabel("Y-as")
ax.set_zlabel("Z-as")

# Toon de plot
plt.show()
```
